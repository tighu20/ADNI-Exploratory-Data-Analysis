```python
#concatinating all the individual csv files
complete_df=pd.concat(dataframe_list).reset_index(drop=True)
complete_df=complete_df.drop_duplicates(subset=['Image Data ID'], keep='first')
```

```python
import os, fnmatch,sys
import shutil 
from os import path
def find(pattern, path):
    result = []
    for root, dirs, files in os.walk(path):
        for name in files:
            if fnmatch.fnmatch(name, pattern):
                result.append(os.path.join(root, name))
    print(len(result))
    return(result)
final_dir = '/pghbio/dbmi/batmanlab/tighu/downloads/ADNI1/unzipped_ADNI/consolidated/by_subjects'
final_image_locations=find('*.nii', final_dir)
image_ids=[]

for location in final_image_locations:
    imgid=location.split("_")[-1]
    imgid = imgid[:-4]
    image_ids.append(imgid)
```
