```python
import os, fnmatch,sys
import shutil 
from os import path
dict_image_locations={}
folder_count,image_count=1,1
counter=1

main_dir = '/pghbio/dbmi/batmanlab/tighu/downloads/ADNI1/unzipped_ADNI/consolidated'
folder_paths=[ "/pghbio/dbmi/batmanlab/tighu/downloads/ADNI1/unzipped_ADNI/consolidated/ADNI1_Complete_3Yr_1.5T",
"/pghbio/dbmi/batmanlab/tighu/downloads/ADNI1/unzipped_ADNI/consolidated/ADNI1_Complete_3Yr_3T",
"/pghbio/dbmi/batmanlab/tighu/downloads/ADNI1/unzipped_ADNI/consolidated/ADNI1_Annual_2_Yr_3T",
"/pghbio/dbmi/batmanlab/tighu/downloads/ADNI1/unzipped_ADNI/consolidated/ADNI1_Complete_2Yr_1.5T",
"/pghbio/dbmi/batmanlab/tighu/downloads/ADNI1/unzipped_ADNI/consolidated/ADNI1_Complete_2Yr_3T",
"/pghbio/dbmi/batmanlab/tighu/downloads/ADNI1/unzipped_ADNI/consolidated/ADNI1_Complete_1Yr_3T",
"/pghbio/dbmi/batmanlab/tighu/downloads/ADNI1/unzipped_ADNI/consolidated/ADNI1_Complete_1Yr_1.5T",
"/pghbio/dbmi/batmanlab/tighu/downloads/ADNI1/unzipped_ADNI/consolidated/ADNI1_Baseline_3T",
"/pghbio/dbmi/batmanlab/tighu/downloads/ADNI1/unzipped_ADNI/consolidated/ADNI1_Screening_1.5T"]

def find(pattern, path):
    result = []
    for root, dirs, files in os.walk(path):
        for name in files:
            if fnmatch.fnmatch(name, pattern):
                result.append(os.path.join(root, name))
    print(len(result))
    return(result)



for path in folder_paths:
    
    dict_image_locations[path.split("/")[-1]]=find('*.nii', path)
    

initial_copy_locations=dict_image_locations['ADNI1_Complete_3Yr_1.5T']+dict_image_locations['ADNI1_Complete_3Yr_3T']


for image_path in initial_copy_locations:
    
    
    image_path_list=image_path.split("/")
    #if subject exists
    if os.path.isdir(main_dir+'/by_subjects/'+image_path_list[11]):
        
        #if year exists
        if os.path.isdir(main_dir+'/by_subjects/'+image_path_list[11]+'/'+image_path_list[13]):
            shutil.copyfile(image_path, main_dir+'/by_subjects/'+image_path_list[11]+'/'+image_path_list[13]+'/'+image_path_list[-1])        
        
        #if year not exists               
        else:
            os.mkdir(main_dir+'/by_subjects/'+image_path_list[11]+'/'+image_path_list[13])            
            shutil.copyfile(image_path, main_dir+'/by_subjects/'+image_path_list[11]+'/'+image_path_list[13]+'/'+image_path_list[-1])
    
    #if subject folder does not exists  
    else:
        os.mkdir(main_dir+'/by_subjects/'+image_path_list[11]+'/')
        os.mkdir(main_dir+'/by_subjects/'+image_path_list[11]+'/'+image_path_list[13]+'/')
        shutil.copyfile(image_path, main_dir+'/by_subjects/'+image_path_list[11]+'/'+image_path_list[13]+'/'+image_path_list[-1])
    print("initial transfer",counter,"done")
    counter=counter+1
    
    
print("initial copy done")
```

```python
del dict_image_locations['ADNI1_Complete_3Yr_1.5T']
del dict_image_locations['ADNI1_Complete_3Yr_3T']
```

```python
folder_count,image_count=3,1
for folder in dict_image_locations.keys():
    
    for image_path in dict_image_locations[folder]:
        print("Currently processing",folder_count,"folder's",image_count,"image")
        image_path_list=image_path.split("/")
        
        #if subject folder exists
        if os.path.isdir(main_dir+'/by_subjects/'+image_path_list[11]):
                
             #if year exists
            if os.path.isdir(main_dir+'/by_subjects/'+image_path_list[11]+'/'+image_path_list[13]):
                #do nothing
                print("No action needed")
            
             #if  year does not exists    
            else:
                os.mkdir(main_dir+'/by_subjects/'+image_path_list[11]+'/'+image_path_list[13])
                shutil.copyfile(image_path, main_dir+'/by_subjects/'+image_path_list[11]+'/'+image_path_list[13]+'/'+image_path_list[-1])
                
                
        
        #else if subject folder does not exist
        else:
            #print(path+'/by_subjects/'+image_path_list[11]+'/')
            os.mkdir(main_dir+'/by_subjects/'+image_path_list[11]+'/')
            os.mkdir(main_dir+'/by_subjects/'+image_path_list[11]+'/'+image_path_list[13]+'/')
            shutil.copyfile(image_path, main_dir+'/by_subjects/'+image_path_list[11]+'/'+image_path_list[13]+'/'+image_path_list[-1])
            
        image_count=image_count+1
        
        
    folder_count=folder_count+1
```
